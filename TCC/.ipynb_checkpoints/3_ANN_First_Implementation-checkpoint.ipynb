### Activation function:

4 main types
* Threshold Function { x < 0 => 0 / x >= 0 => 1}
* Sigmoid Function { f(x) = 1 / 1 + e^-x    => from 0 to 1 }
* Rectifier { f(x) = max(x,0) => from 0 to inf }
* Hyperbolic Tangent (tanh) { f(x) = 1 - e^-2x / 1+e^-2x => from -1 to 1 }
